```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

## V1674HERB

```python
date = 20
freq = 4.68e9

conv = (3600)**(-1)*(np.pi/180) #mas to rad
jy = 10e-26

theta1 = 3.417e-02 *conv
theta2 = 1.515e-02 *conv

flux =  9.7775e-04*jy



k = 1.38e-23
c = 2.99e8
```

```python
Tb = (c/freq)**2 * (flux) * (2*k)**(-1) * ((4*np.log10(2))/(np.pi*theta1*theta2))
```

```python
f'{Tb:.2E}'
```

## V1674HERD

```python
date = 24
theta2d = 1.515e-02 *conv
theta1d = 3.417e-02 *conv
fluxd = 1.2548e-03*jy



Tbd = (c/freq)**2 * (fluxd) * (2*k)**(-1) * ((4*np.log10(2))/(np.pi*theta1d*theta2d))

f'{Tbd:.2E}'
```
