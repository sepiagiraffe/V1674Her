```python
import matplotlib.pyplot as plt
import pandas as  pd
# import numpy as np
from matplotlib.ticker import FormatStrFormatter
```

```python
data = pd.read_csv("./aavsodata_665626de4df53.txt");
# df.loc[df['column_name'] == some_value]
# df['Prices'] = pd.to_numeric(df['Prices'], errors='coerce')
data = data.loc[data['Validation Flag'] == "Z"];
data['Magnitude'] = pd.to_numeric(data['Magnitude'], errors='coerce');
data_b = data.loc[data['Band'] == "B"];
data_v = data.loc[data['Band'] == "V"];
# data_v = data_v.loc[data['Observer Affiliation'] == "AAVSO"];
```

```python
plt.style.use('presentation');
fig, ax = plt.subplots(figsize=(21,11));

ax.xaxis.set_major_formatter(FormatStrFormatter('%4.2e'));
ax.scatter(data_b['JD'], data_b['Magnitude'], color='blue');
ax.scatter(data_v['JD'], data_v['Magnitude'], color='lime');
ax.vlines(2460018.5000000, ymax=12.25, ymin=8.5);
ax.invert_yaxis();
ax.set_title('T CrB Light Curve from 2023 - Now');
ax.set_xlabel('Julian Date');
ax.set_ylabel('Magnitude');
```
